## 6.requests

### 基本get请求

```python
import requests
r = requests.get('http://httpbin.org/get')
print(r.url)
```

```python
payload = {'key1': 'value1','key2': 'value2'}
r = requests.get('http://httpbin.org/get', params = payload)
print(r.url)
```

```python
headers = {'hello': 'world'}
r = requests.get('http://httpbin.org/get', headers = headers)
print(r.text)
```

### 基本post请求

```python
payload = {'hello': 'world'}
r = requests.post('http://httpbin.org/post', data = payload)
print(r.text)
```

#### 发送json格式数据

```python
import json
payload = {'hello': 'world'}
r = requests.post('http://httpbin.org/post', data = json.dumps(payload))
print(r.text)
```

#### 上传文件

```python
url = 'http://httpbin.org/post'
files = {'file': open('baidu.txt', 'rb')}
r = requests.post(url, files = files)
print(r.text)
```

#### Cookies

```python
url = 'http://httpbin.org/cookies'
cookies = dict(cookies_are = 'working')
r = requests.get(url, cookies = cookies)
print(r.text)
```

### 请求超时配置

```python
url = 'http://github.com'
r = requests.get(url, timeout = 100)
print(r.text)
```

### 持久会话
#### 没有持久会话的情况

```python
r = requests.get('http://httpbin.org/cookies/set/sessioncookie/123456789')
print(r.text)
```

```python
r = requests.get('http://httpbin.org/cookies')
print(r.text)
```

#### 设置持久会话的情况

```python
s = requests.Session()
s.get('http://httpbin.org/cookies/set/sessioncookie/123456789')
r = s.get('http://httpbin.org/cookies')
print(r.text)
```

### 代理

```python
proxies = {'https': 'http//41.118.132.69:4433'}
r = requests.post('https://baidu.com', proxies = proxies)
print(r.status_code)
```

```python
r = requests.post('https://baidu.com')
print(r.status_code)
```
