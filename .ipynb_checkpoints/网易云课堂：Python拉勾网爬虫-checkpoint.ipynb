```python
import requests
from bs4 import BeautifulSoup
```

```python
def main():
    url = 'https://www.lagou.com/jobs/list_python?px=default&city=%E5%8C%97%E4%BA%AC#filterBox'
    result = requests.get(url)
    print(result.content)

if __name__ == '__main__':
    main()
```

以上结果是因为拉勾网有反爬虫机制
Chrome→开发者工具→network，刷新网址→Name→list_python?...→Headers→Request Headers，一般检测cookies, host, referer, upgrade-Insecure-Requests,user-agent

```python
def main():
    headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.75 Safari/537.36',
    'Host': 'www.lagou.com',
    'Referer': 'http://www.lagou.com'
    }
    url = 'https://www.lagou.com/jobs/list_python?px=default&city=%E5%8C%97%E4%BA%AC#filterBox'
    result = requests.get(url, headers = headers)
    print(result.content)

if __name__ == '__main__':
    main()
```
