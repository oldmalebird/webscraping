#### 只爬取第一页的职位详细页：headers没加Cookie信息，只能爬取一个

```python
def crawl_detail(id):
    url = 'https://www.lagou.com/jobs/%s.html' %id
    headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.75 Safari/537.36',
    'Host': 'www.lagou.com',
    'Referer': 'https://www.lagou.com/jobs/list_python?px=default&city=%E5%8C%97%E4%BA%AC',
    'Upgrade-Insecure-Requests': '1'
    }
    req = requests.get(url, headers = headers)
    soup= BeautifulSoup(req.content) #soup= BeautifulSoup(req.content, 'lxml')
    job_bt = soup.find_all('dd', attrs = {'class':'job_bt'})
    print(job_bt)
    return job_bt
    
def main():
    headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.75 Safari/537.36',
    'Host': 'www.lagou.com',
    'Referer': 'https://www.lagou.com/jobs/list_python?px=default&city=%E5%8C%97%E4%BA%AC',
    'X-Anit-Forge-Code': '0',
    'X-Anit-Forge-Token': None,
    'X-Requested-With': 'XMLHttpRequest'
    }
    url = 'https://www.lagou.com/jobs/positionAjax.json?px=default&city=%E5%8C%97%E4%BA%AC&needAddtionalResult=false'
    data = {
        'first': 'true',
        'pn' : '1',
        'kd' : 'python'
    }
    result = requests.post(url, headers = headers, data = data)
    json_result = result.json()
    positions = json_result['content']['positionResult']['result']
    for position in positions:
        position_dict = {
            'position_name': position['positionName'],
            'work_year': position['workYear'],
            'salary': position['salary'],
            'district': position['district'],
            'company_name': position['companyFullName']}
        position_id = position['positionId']
        position_detail = crawl_detail(position_id)
        position_dict['position_detail'] = position_detail
        positions.append(position_dict)
    
    print(positions)

if __name__ == '__main__':
    main()
```
